```python
#Import dependency
import pandas as pd
```

```python
#define passing criterion
MIN_PASSING_SCORE=70
```

```python
#Import schools csv file
#check if it transfered correctly
schools="schools_complete.csv"
schools_df=pd.read_csv(schools)
schools_df.head()
```

```python
school_types = schools_df.set_index(["school_name"])["type"]
school_types
```

```python
#Import students csv file
#check if it transfered correctly
students="students_complete.csv"
students_df=pd.read_csv(students)
students_df.head()
```

```python
#Merge datasets - inner merging
schools_complete_df = pd.merge(schools_df,students_df,on="school_name", how="inner")
schools_complete_df.head()
```

```python
#Changing index
schools_complete_df = schools_complete_df.set_index(["school_name"])
schools_complete_df
```

```python
#District summary
#Total number of unique schools
total_schools=len(schools_complete_df["School ID"].unique())
total_schools
```

```python
#District summary
#Total number of students
total_students = len(schools_complete_df["Student ID"].unique())
total_students
```

```python
#District summary
#Total budget
total_budget=schools_complete_df.groupby("school_name").mean()["budget"].sum()
total_budget
```

```python
#District summary
#Average math score
avg_m_score=schools_complete_df['math_score'].mean()
avg_m_score
```

```python
#District summary
#Average reading score
avg_r_score=schools_complete_df['reading_score'].mean()
avg_r_score
```

```python
#District summary
#passing math_count
passing_math=schools_complete_df["math_score"]>=MIN_PASSING_SCORE
passing_math_count=schools_complete_df[passing_math].count()["Student ID"]
passing_math_count
```

```python
#District summary
#% passing math 
perc_math_pass = (passing_math_count/total_students)*100
perc_math_pass
```

```python
#District summary
#passing reading_count
passing_reading=schools_complete_df["reading_score"]>=MIN_PASSING_SCORE
passing_reading_count=schools_complete_df[passing_reading].count()["Student ID"]
passing_reading_count
```

```python
#District summary
#% passing reading
perc_reading_pass = (passing_reading_count/total_students)*100
perc_reading_pass
```

```python
#District summary
#passing math and reading (overall)count
passing_overall=(schools_complete_df["math_score"]>=MIN_PASSING_SCORE) & (schools_complete_df["reading_score"]>=MIN_PASSING_SCORE)
passing_overall_count=schools_complete_df[passing_overall].count()["Student ID"]
passing_overall_count
```

```python
#District summary
#% passing math and reading (overall)
perc_overall_pass = (passing_overall_count/total_students)*100
perc_overall_pass
```

```python
#district summary-snapshot of district's key metrics
#creating columns and defnining each column 
district_summary= [{"Totaluniqueschools":total_schools, "Totalstudents":total_students,"Totalbudget":total_budget,"Averagemathscore":avg_m_score,"Averagereadingscore":avg_r_score,"mathpass%":perc_math_pass,"readingpass%":perc_reading_pass,"overallpassing%":perc_overall_pass}]
district_summary_df = pd.DataFrame(district_summary)
district_summary_df
```

```python
#district summary
#formatting columns
district_summary_df["Totalstudents"]= district_summary_df["Totalstudents"].map("{:,}".format)
district_summary_df["Totalbudget"]= district_summary_df["Totalbudget"].map("${:,.1f}".format)
district_summary_df["Averagemathscore"]=district_summary_df["Averagemathscore"].map("{:.2f}".format)
district_summary_df["Averagereadingscore"]=district_summary_df["Averagereadingscore"].map("{:.2f}".format)
district_summary_df["mathpass%"]= district_summary_df["mathpass%"].map("{:.2f}%".format)
district_summary_df["readingpass%"]= district_summary_df["readingpass%"].map("{:.2f}%".format)
district_summary_df["overallpassing%"]= district_summary_df["overallpassing%"].map("{:.2f}%".format)
district_summary_df
```

```python
#School summary
#Total students per school
total_students_perschool=schools_complete_df.groupby("school_name").count()["student_name"]
total_students_perschool
```

```python
#School summary
#school type
schools_types=len(schools_complete_df["type"].unique())
schools_types
```

```python
#School summary
#Total budget per school
budget_perschool=schools_complete_df.groupby("school_name").mean()["budget"]
budget_perschool
```

```python
#start creating dataframe grouped byschool
school_summary = pd.merge(total_students_perschool,budget_perschool,on="school_name")
school_summary
```

```python
#add school types to school_summary dataframe
school_summary_df = pd.merge(school_types,school_summary,on="school_name", how="inner")
school_summary_df
```

```python
#Rename column
school_summary_df = school_summary_df.rename(columns={'student_name': 'total_students'})
school_summary_df
```

```python
#School summary
#Per_student budget per school
school_summary_df["budget_perstudent"]=(school_summary_df["budget"]/school_summary_df["total_students"])
school_summary_df
```

```python
#School summary
#Average math score
avgmathscore_perschool=schools_complete_df.groupby("school_name").mean()["math_score"]
avgmathscore_perschool

```

```python
#Add average mathscore column to school summary dataframe
school_summary_df["avgmathscore"]=avgmathscore_perschool
school_summary_df
```

```python
#School summary
#Average reading score
#Add average reading score column to school summary dataframe
avgreadingscore_perschool=schools_complete_df.groupby("school_name").mean()["reading_score"]
avgreadingscore_perschool
school_summary_df["avgreadingscore"]=avgreadingscore_perschool
school_summary_df
```

```python
#School summary
#passing math_count
passing_math=schools_complete_df.loc[schools_complete_df["math_score"]>=MIN_PASSING_SCORE]
passing_math
```

```python
#School summary
# Number of students per school passed math
count_mathpass=passing_math.groupby('school_name')['math_score'].count()
count_mathpass
```

```python
#School summary
# Adding passing math_count to school summary df
school_summary_df["passmathcount"]=count_mathpass
school_summary_df
```

```python
#School summary
#% passing math
passingmathperc = school_summary_df["passmathcount"] /school_summary_df["total_students"]*100 
passingmathperc

```

```python
# Adding % passing math to school summary df
school_summary_df["mathpassperc"]=passingmathperc
school_summary_df
```

```python
#School summary - reading
#passing reading count
passing_reading=schools_complete_df.loc[schools_complete_df["reading_score"]>=MIN_PASSING_SCORE]
passing_reading
# Number of students per school passed reading
count_readpass=passing_reading.groupby('school_name')['reading_score'].count()
count_readpass
# Adding passing reading count to school summary df
school_summary_df["passreadingcount"]=count_readpass
school_summary_df
#% passing reading
passingreadingperc = school_summary_df["passreadingcount"] /school_summary_df["total_students"]*100 
passingreadingperc
# Adding % passing reading to school summary df
school_summary_df["readpassperc"]=passingreadingperc
school_summary_df
```

```python
#School summary - overall - reading and math
#passing overall count
passing_overall=(schools_complete_df["math_score"]>=MIN_PASSING_SCORE) & (schools_complete_df["reading_score"]>=MIN_PASSING_SCORE)
passing_overall
# Number of students per school passed overall
#schools_complete_df[passing_overall].count()["Student ID"]
count_overallpass=passing_overall.groupby("school_name").sum()
count_overallpass
# Adding overall passing count to school summary df
school_summary_df["overallpasscount"]=count_overallpass
school_summary_df
#% passing overall 
passingoverallperc = school_summary_df["overallpasscount"] /school_summary_df["total_students"]*100 
passingoverallperc
# Adding % passing overall to school summary df
school_summary_df["overallpassperc"]=passingoverallperc
school_summary_df
```

```python
#Highest-Performing Schools (by % Overall Passing)
#display top 5 rows
top_schools_df= school_summary_df.sort_values(["overallpassperc"],ascending = False)
top_schools_df.head()
```

```python
#Lowest-Performing Schools (by % Overall Passing)
#display top 5 rows
bottom_schools_df= school_summary_df.sort_values(["overallpassperc"])
bottom_schools_df.head()
```

```python
#Avg Math Scores by Grade at each school
#Define values of each grade
ninth_graders = schools_complete_df[(schools_complete_df["grade"] == "9th")]
tenth_graders = schools_complete_df[(schools_complete_df["grade"] == "10th")]
eleventh_graders = schools_complete_df[(schools_complete_df["grade"] == "11th")]
twelfth_graders = schools_complete_df[(schools_complete_df["grade"] == "12th")]
```

```python
# Create a copy of the school summary since we are going to categorize it on base of budget per student 
school_spending_df = school_summary_df.copy()
school_spending_df
```

```python
#categorize schools on basis of budget per student
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Add column  average spending ranges (per student) to the school summary database
school_spending_df["avg_spendingrange"] = pd.cut(school_spending_df["budget_perstudent"], spending_bins, labels=labels)
school_spending_df                                                              
```

```python
#mean scores per spending range
Avgmath_scores = school_spending_df.groupby(["avg_spendingrange"]).mean()["avgmathscore"]
Avgreading_scores = school_spending_df.groupby(["avg_spendingrange"]).mean()["avgreadingscore"]
Avgpassing_math = school_spending_df.groupby(["avg_spendingrange"]).mean()["mathpassperc"]
Avgpassing_reading = school_spending_df.groupby(["avg_spendingrange"]).mean()["readpassperc"]
Avgoverall_passing = school_spending_df.groupby(["avg_spendingrange"]).mean()["overallpassperc"]
```

```python
#create new dataframe spending_summary_df
spending_summary_df=school_spending_df[["avg_spendingrange"]]
spending_summary_df
```

```python
#Add mean scores per spending range to spending_summary_df
spending_summary_df = pd.DataFrame ( { "Average Math Score" : Avgmath_scores,
                                              "Average Reading Score" : Avgreading_scores,
                                              "% Passing Math" : Avgpassing_math, "% Passing Reading" : Avgpassing_reading, "% Overall Passing" : Avgoverall_passing,
                                          }) 
spending_summary_df
```
